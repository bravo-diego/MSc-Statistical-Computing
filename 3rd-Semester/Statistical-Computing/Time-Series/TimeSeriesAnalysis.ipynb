# Time Series Analysis

## Stochastic Processes

A stochastic process is a family of random variables $\{X_{t}, t \in T\}$ defined on a probability space ($\Omega, \mathcal{F}, \mathcal{P}$). The functions $\{X, (\omega), \omega \in \Omega\}$ on T are known as the realizations or sample-paths of the process $\{X_{t}, t \in T\}$. In other words, a **realization** refers to one specific outcome or sample path of the process over time. Since stochastic processes involve randomness and evolve over time, each realization represents the behaviour of the process for one particular sequence of random events.

A **time series** is a set of observations $x_{t}$, each one being recorded at a specified time $t$. The term time series is usually used to refer both the data and the process of which it is a realization.

**Kolmogorov's Theorem**. To define a stochastic process one approach is to specify all its finite-dimensional distributions. If we have a collection of such finite-dimensional distributions that are consistent with each other, then there exists a stochastic process that has these finite-dimensional distributions.

## Stationarity and Strict Stationarity

A time series $X(t)$ is said to be **weakly stationary** if:

- The expected value of the process is constant over time.

$$ 
E[X_{t}] = E[X_{t + k}] = \mu \quad \text{for all $t, k$}
$$

- The variance of the process is constant over time.

$$ 
Var(X_{t}) = Var(X_{t + k}) = \sigma^{2} \quad \text{for all $t, k$}
$$

- The autocovariance between two time points $t$ and $t + k$ depends only on the time lag $k$, not on the actual time points $t$ and $t + k$.

$$ 
\gamma_{X}(r, s) = \gamma_{X}(r + t, s + t) \quad \text{for all $r, s, t \in Z$},
$$

$\qquad$ if we know the autocovariance between $X_r$ and $X_s$, we can confidently say that the same autocovariance holds between $X_{r + t}$ and $X_{s + t}$ after shifting the entire time series forward by $t$ units.

**Autocovariance Function**. If $\{X_{t}, t \in T\}$ is a process such that $Var(X_{t}) < \infty$ for each $t \in T$, then the autocovariance function $\gamma_{X}(.,.)$ of $\{X_t\}$ is defined as follows:

$$
\gamma_{X}(r, s) = Cov(X_{r}, X_{s}) = E[(X_r - \mu_r)(X_s - \mu_s)] \quad r, s, \in T
$$

The autocovariance function provides insights into the correlation structure of the process, how the random variables within a time series are linearly related to each other across different points in time.

**Autocorrelation Function**. The autocorrelation function of $\{X_{t}\}$ is defined as:

$$
\rho_{X}(h) = \frac{\gamma_x(h)}{\gamma_x(0)} = Corr(X_{t + h}, X_{t}) \quad \text{for all $t, h \in Z$}
$$

$\qquad$ where $\gamma(0) = Var(X_t) = E[(X_t - \mu)^2]$.

The autocorrelation function measures the correlation between a time series and a lagged version of itself. It provides insights into the relationship between observations at different time points. 

The time series $X(t)$ is said to be **strictly stationary** if the joint distributions of $(X_{t1}, ..., X_{tk})$ and $(X_{t1 + h}, ..., X_{tk + h})$ are the same for all positive integers $k$ and for all $t_{1}, ..., t_{k}, h \in Z$. 

Strict stationarity means intuitively that graphs over two equal-length time intervals of a realization of the time series shoul exhibit similar statistical characteristics. This is a stronger condition than weakly stationarity sice it requires the entire distribution to be invariant under time shifts, not just the first two moments.

## Moving Average Process (MA)

A moving average process is a type of time series model that expresses the current value of the time series as a linear combination of actual and past forecast errors, also called **shocks**. The term *shocks* refers to an unexpected or sudden change in a variable that influences the value of a time series. Shocks are often modeled as a random disturbances or noise that can affect the current and future values of the time series. 

In many models, shocks are assumed to be white noise, i.e. they have a mean of zero, constant variance and no autocorrelation. This assumption simplifies the analysis and modeling of time series data. 

The process $\{Z_t\}$ is said to be white noise with mean $0$ and variance $\sigma^{2}$:

$$
\{Z_t\} \sim WN(0, \sigma^2)
$$

In a **MA(q)** process, the current value of the series is influenced by the current shock and past shocks:

$$
Y_t = Z_{t} + \theta_{1}Z_{t - 1} + ... + \theta_{q}Z_{t - q} \qquad Z_{t} \sim WN(0, \sigma^2),
$$

$\qquad$ where $\theta_{q}$ are the coefficients of the lagged error terms and $q$ is the order of the MA. 

By using $Z_t$ we can explicitly denote that the error terms are uncorrelated random variables with a specific distribution. 

$$
\epsilon_t \sim WN(0, \sigma^{2})
$$

A **MA(1)** process ($q = 1$) is expressed as:

$$
Y_t = \epsilon_t + \theta\epsilon_{t - 1} \qquad \epsilon_t \sim WN(0, \sigma^2), 
$$

here the current value $Y_t$ depends on the current noise $\epsilon_t$ and the immediately preceding noise $\epsilon_{t - 1}$. 

MA processes have a **finite memory** effect, meaning that the current value of the time series is influenced only by the most recent $q$ errors. Errors beyond these $q$ lags have no direct effect on the current value

The *characteristic equation* of a MA process is a polynomial equation derived from the coefficients of the model. This equation is central to determining whether the process is invertible, as the roots of the characteristic equation must satisfy certain conditions for invertibility. 

Consider a general MA(q) process:

$$
Y_t = Z_{t} + \theta_{1}Z_{t - 1} + \theta_{2}Z_{t - 2} + ... + \theta_{q}Z_{t - q},
$$

the characteristic equation is obtained by expressing the MA process in terms of the backshift (also called lag) operator $B$ where $B^{k}X_{t} = X_{t - k}$, thus:

$$
Y_t = Z_{t} + \theta_{1}BZ_t + \theta_{2}B^{2}Z_t + ... + \theta_{q}B^{q}Z_t
$$

$$
= (1 + \theta_{1}B + \theta_{2}B^{2} + ... + \theta_{q}B^{q})Z_{t}
$$

The characteristic equation is the equation obtained by setting the polynomial equal to zero:

$$
1 + \theta_{1}B + \theta_{2}B^{2} + ... + \theta_{q}B^{q} = 0
$$

For a MA(1) process the characteristic equation is:

$$
1 + \theta_{1}B = 0
$$

Solving for B:

$$
B = -\frac{1}{\theta_{1}}
$$


So for a MA process to be invertible, the roots of the characteristic equation must lie outside the unit circle in the complex plane. This implies:

$$
\vert \theta_{1} \vert < 1
$$

A MA process is considered **invertible** when it can be expressed equivalently as an infinite-order AR (autoregressive) process. Invertibility allowss to express the current valuee of the time series as a function of past values of the series itself, rather than as a function of past shocks or errors.

### Verification of Model Residual Diagnostics

If all systematic information has been extracted using a forecasting model, what remains—the residual—should be white noise. More precisely, true innovations are white noise, and if a model is a good approximation of Wold's theorem, then its one-step-ahead forecast errors should approximate white noise.

The model residuals are the in-sample analog of the out-of-sample one-step-ahead forecast errors. Consequently, we see the utility of several hypothesis tests to assess whether the residuals are white noise. The Durbin-Watson test is the most commonly used.

Recall that the Durbin-Watson statistic, described in the appendix of Chapter 1, is:

$$
DW = \frac{\sum_{t = 2}^{T}(e_t - e_{t - 1})^{2}}{\sum_{t = 1}^{T}e^{2}_{t}}
$$

Observe that:

$$
\sum_{t = 2}^{T}(e_t - e_{t - 1})^{2} \sim 2\sum_{t = 2}^{T} e_{t}^{2} - 2\sum_{t = 2}^{T}e_{t}e_{t - 1}
$$

and thus:

$$
DW = 2(1 - \hat{p}(1))
$$

The Durbin-Watson test is effectively based solely on the first-order correlation and essentially tests whether the first autocorrelation is zero. Consequently, the Durbin-Watson test is referred to as a **first-order serial correlation test**, or a test for first-order serial correlation. Furthermore, the Durbin-Watson test is not valid in the presence of lagged dependent variables. In such cases, we prefer a more general and flexible framework for diagnosing serial correlation. The residual correlogram, comprising the sample residual autocorrelations, partial sample autocorrelations, and associated $Q$ statistics, fulfills this role.


a) When describing the correlogram in the test, we focus on the case of an observed time series, for which we demonstrate that the $Q$ statistics are distributed as $\chi_m^2$. However, we now wish to evaluate whether the unobserved disturbances of the model are white noise. To do so, we use the model residuals, which are estimates of the unobserved disturbances. Since a model is fitted to obtain the residuals, we need to account for the degrees of freedom used. The consequence is that the distribution of the $Q$ statistic under the white noise hypothesis is better approximated as a $\chi_{m-k}^2$ random variable, where $k$ is the number of parameters estimated. This is why, for example, $p$-values are not mentioned (in fact, statistical programs do not even calculate them) for the $Q$ statistics associated with the residual correlogram of our employment forecasting model until $m > k$.

### **Durbin's H Test**

The Durbin-Watson statistic, developed by econometricians Durbin & Watson, was designed to detect significant first-order autocorrelation in the residuals of a regression analysis. Its efficiency in contexts assuming independence in linear regression models led to its widespread use. However, this resulted in its application in inappropriate situations, such as cases involving **lagged dependent variables**, leading to erroneous conclusions. To address these cases, Durbin (1970) proposed the $h$ test, which allows for evaluating **autocorrelation in first-order autoregressive processes**.

The $h$ statistic is defined as:

$$
h = \left(1 - \frac{1}{2}d\right)\left[\frac{n}{1 - n\hat{V}(\hat{\beta_{1}})} \right]^{\frac{1}{2}}
$$

$\qquad$ where $d$ is the Durbin-Watson statistic, $n$ is the sample size, and $\hat{V}$ represents the estimated variance of $\hat{\beta_{1}}$. Here, $\beta_{1}$ is the coefficient of the lagged dependent variable, and $\hat{\beta_{1}}$ is its estimate.

Durbin demonstrated that, under certain assumptions, $h$ asymptotically follows a normal distribution with zero mean and unit variance (i.e., $h \sim N(0, 1)$) when the errors are uncorrelated. Furthermore, he suggested using the standard normal distribution table to test the null hypothesis of uncorrelated errors.


**References**

- Park, S.-B. (1975). On the small-sample power of Durbin's h test. Journal of the American Statistical Association, 70(349), 60-63. 
- Proïa, F. (2013). Further results on the h-test of Durbin for stable autoregressive processes. Journal of Multivariate Analysis, 118, 77-101.

### **Breusch-Godfrey Test**

Breusch & Godfrey (1978) proposed the use of Lagrange Multiplier (LM) tests to detect autocorrelation in the residuals of a regression model. This method is based on the concept of 'locally equivalent alternatives' (LEA), which allows for identifying even mild autocorrelations in the errors. In other words, the test is designed to detect small deviations from independence in the model's residuals.

The method involves applying two regressions. The first, called the primary regression, is performed on the original model. Then, the residuals estimated from this regression are used as the dependent variable in a second regression, known as the auxiliary regression, where the independent variables include the regressors from the original model plus the lagged residuals (past residual values).

The test statistic is obtained by multiplying the number of observations by the coefficient of determination from the auxiliary regression. Thus, the statistic $TR^{2}$ follows a $\chi^{2}$ distribution under the null hypothesis ($H_{0}$: independence of the residuals).


The Breusch-Godfrey test has the advantage of being valid for dynamic models and can be easily extended to detect **higher-order autocorrelation**.

**References**

- Edgerton, D. & Shukur, G. (1999) Testing autocorrelation in a
system perspective testing autocorrelation, Econometric Reviews, 18:4, 343-386,

***

The **Breusch-Godfrey test** is considered more useful due to its ability to identify autocorrelation across multiple lags ($t − 1, t − 2,$ etc.). In contrast, the **Durbin $h$ test** is limited to the residuals of the immediately preceding periods ($t$ and $t−1$). On the other hand, the **residual correlogram** could be a very valuable tool for visually displaying the autocorrelation of the residuals and serves as an initial step before applying a more formal statistical test for evaluating the residuals of a model.

In summary, each of these tests has its own importance and utility, depending on the complexity of the problem and the objectives being pursued.

### Exercise 2.0

Prove step by step that:

$$
\gamma(\tau) = E(y_{t}y_{t - \tau}) = E((\epsilon_{\tau} + \theta\epsilon_{t - 1})(\epsilon_{t - \tau} + \theta\epsilon_{t - \tau - 1})) 
\begin{cases} 
\theta\sigma^{2}, & \tau = 1 \\
0, & \text{otherwise} 
\end{cases}
$$

By completing the missing steps and explicitly evaluating the expectation $E((\epsilon_{\tau} + \theta\epsilon_{t - 1})(\epsilon_{t - \tau} + \theta\epsilon_{t - \tau - 1}))$.

Expanding the expression:

$\begin{aligned}
\gamma(\tau) = E(y_{t}y_{t - \tau}) &= E((\epsilon_{t} + \theta\epsilon_{t - 1})(\epsilon_{t - \tau} + \theta\epsilon_{t - \tau - 1}))\\
&= E(\epsilon_{t}\epsilon_{t - \tau} + \theta\epsilon_{t}\epsilon_{t - \tau - 1} + \theta\epsilon_{t - 1}\epsilon_{t - \tau} + \theta^{2}\epsilon_{t - 1}\epsilon_{t - \tau - 1})\\
&= E(\epsilon_{t}\epsilon_{t - \tau}) + E(\theta\epsilon_{t}\epsilon_{t - \tau - 1}) + E(\theta\epsilon_{t - 1}\epsilon_{t - \tau}) + E(\theta^{2}\epsilon_{t - 1}\epsilon_{t - \tau - 1})
\end{aligned}$

$\qquad$ where $\epsilon_{t} \sim WN(0, \sigma^{2})$.

Let $\tau = 1$.

$\begin{aligned}
E(y_{t}y_{t - 1}) &= E(\epsilon_{t}\epsilon_{t - 1}) + E(\theta\epsilon_{t}\epsilon_{t - 1 - 1}) + E(\theta\epsilon_{t - 1}\epsilon_{t - 1}) + E(\theta^{2}\epsilon_{t - 1}\epsilon_{t - 1 - 1})\\
&= E(\epsilon_{t}\epsilon_{t - 1}) + E(\theta\epsilon_{t}\epsilon_{t - 2}) + E(\theta\epsilon^{2}_{t - 1}) + E(\theta^{2}\epsilon_{t - 1}\epsilon_{t - 2})\\
&= E(\epsilon_{t})E(\epsilon_{t - 1}) + \theta E(\epsilon_{t})E(\epsilon_{t - 2}) + \theta E(\epsilon^{2}_{t - 1}) + \theta^{2}E(\epsilon_{t - 1}) E(\epsilon_{t - 2})\\
&= \theta E(\epsilon^{2}_{t - 1})\\
&= \theta \sigma^{2}
\end{aligned}$

Therefore, when $\tau = 1$:

$$
E(y_{t}y_{t - \tau}) = \theta \sigma^{2}
$$

Let $\tau > 1$.

$\begin{aligned}
E(y_{t}y_{t - i}) &= E(\epsilon_{t}\epsilon_{t - i}) + E(\theta\epsilon_{t}\epsilon_{t - i - 1}) + E(\theta\epsilon_{t - 1}\epsilon_{t - i}) + E(\theta^{2}\epsilon_{t - 1}\epsilon_{t - i - 1})\\
&= E(\epsilon_{t}\epsilon_{t - i}) + E(\theta\epsilon_{t}\epsilon_{t - k}) + E(\theta\epsilon_{t - 1}\epsilon_{t - i}) + E(\theta^{2}\epsilon_{t - 1}\epsilon_{t - k})\\
&= E(\epsilon_{t})E(\epsilon_{t - i}) + \theta E(\epsilon_{t})E(\epsilon_{t - k}) + \theta E(\epsilon_{t - 1})E(\epsilon_{t - i}) + \theta^{2}E(\epsilon_{t - 1})E(\epsilon_{t - k})\\
&= 0
\end{aligned}$

$\qquad$ $i > 1$.

Thus, when $\tau > 1$:

$$
E(y_{t}y_{t - \tau}) = 0
$$

### Model Comparison

In this analysis, both models, an aggregated model and a disaggregated model, were applied to the dataset.

In the case of the **aggregated model**, one of its main advantages is that it is easier to interpret and manage. Additionally, this approach allows for identifying general trends at the macro level, such as seasonal patterns or changes in demand. However, one disadvantage of the aggregated model is the loss of detail, as it does not allow for exploring specific differences between items, which is crucial for making more accurate and informed decisions.

On the other hand, the **disaggregated model** offers the ability to identify patterns, changes, or anomalies at a deeper level, allowing for more detailed analysis of specific products.

For this dataset, the ACF plot of the disaggregated model shows a higher number of peaks at different lags and presents higher autocorrelation values compared to the aggregated model. This suggests a greater complexity in the structure of the disaggregated data. Therefore, it can be concluded that the **aggregated model is more suitable** for making predictions in this case, as it simplifies autocorrelation and improves predictive capability.

### Conclusion

The **aggregated model** is useful when seeking general trends and prioritizing a more "simple" analysis, especially when variability at a deeper level of detail is not very relevant. On the other hand, the **disaggregated model** is recommended if a deep analysis is needed to capture significant variations between specific products.
